**Graphical Representation**

Bar Graph:

```python
plt.subplot(221)
df['Ship Mode'].value_counts().plot(kind='bar', title='Ship Mode', figsize=(16,9))
plt.xticks(rotation=0)

plt.subplot(222)
df['Segment'].value_counts().plot(kind='bar', title='Segment', figsize=(16,9))
plt.xticks(rotation=0)

plt.subplot(223)
df['Category'].value_counts().plot(kind='bar', title='Category', figsize=(16,9))
plt.xticks(rotation=0)

plt.subplot(224)
df['Sub-Category'].value_counts().plot(kind='bar', title='Sub-Category', figsize=(16,9))
plt.xticks(rotation=90)
```

We can conclude many info fron these graphs. People prefer standard class ship mode then rest of the modes. Office supplies are more purchased then others, etc.

Box Plot:

```python
df.boxplot()
```

A boxplot helps us in visualizing the data in terms of quartiles. It also identifies outliers in the dataset, if any.The horizontal green line in the middle represents the median of the data. The hollow circles near the tails represent outliers in the dataset.

Here we can see much of outliers in Sales as there is possiblity of increases or decrease in sale.

Distribution Plot:

```python
sns.distplot(df["Sales"])
```

```python
fig, ax = plt.subplots(figsize=(15,5))
sns.distplot(df['Profit'])
```

 A distribution plot tells us how the data is distributed.

```python
fig, ax = plt.subplots(figsize=(25,5))
#colour=['red' if x < 0 else 'green' for x in df['Profit']]
plt.plot(df['Profit'])
```

```python
corr = df.corr()
sns.heatmap(corr,annot = True, square=True)
plt.yticks(rotation=0)
```

```python
axes,fig=plt.subplots(0,1,figsize=(15,5))
sns.scatterplot("Discount",'Profit',data=df)
```

Here it can be observed that when the discount is till 0.3, there is a profit.

But if the discount increases beyond 0.3 there is a loss happening

```python
sns.scatterplot("Sales",'Profit',data=df)
```

There is no necesserly increase in profit with the increase in sale.

```python
sns.lmplot(x='Profit', y='Sales',data=df,fit_reg=False,col='Category')
```

The profit high in Technology sector and very low almost 0 in Furniture.

```python
sns.lmplot(x='Profit',y='Sales',data=df,fit_reg=False,col='Ship Mode')
```

Profit is high in Standard Class Ship mode, while less in Same day.

```python
sns.lmplot(x='Profit',y='Sales',data=df,fit_reg=False,col='Segment')
```

Profit is high in Consumer segment.

**Conclusion:**

* When the discount is till 0.3, there is a profit so, the discount should not be increased beyond 0.3.

* THE FURNITURE CATEGORY IS CAUSING A LOT OF LOSS, SO THE COMPANY CAN STOP SELLING FURNITURES OR SHOULD INCREASE THE PRICE OF THE FURNITURE CATEGORY

* The profit is high when the ship mode is "Standard Class" and the Profit is negligible when the ship Mode is "Same day" so, the shipping charges for "Same day" should be increased.

* THE COMPANY SHOULD FOCUS MORE ON THE TECHNOLOGY SECTION.
